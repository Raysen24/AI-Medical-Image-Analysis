```python
def normalize_conclusion(conclusion):
    # Define any additional normalizations you want to perform
    normalization_rules = {
        'ileus paralitik atau ileus obstruksi': 'ileus obstruktif atau paralitik',
        'degenerative disc, spondyloarthritis': 'spondyloarthritis, degenerative disc',
        'spondylolisthesis, spondyloarthritis': 'spondyloarthritis, spondylolisthesis',
        'scoliosis, straight lumbal, spondyloarthritis, degenerative disc': 'straight lumbal, spondyloarthritis, scoliosis, degenerative disc',
        'fraktur, straight lumbal, spondyloarthritis': 'straight lumbal, fraktur, spondyloarthritis',
        'hernia nucleus pulposus, straight lumbal, spondyloarthritis': 'straight lumbal, spondyloarthritis, hernia nucleus pulposus',

        'straight cervical, spondyloarthritis, kalsifikasi': 'straight cervical, spondyloarthritis',
        'spondylitis, degenerative disc': 'spondylitis',
        'kifosis, spondyloarthritis, degenerative disc': 'spondyloarthritis',
        'scoliosis, straight lumbal, fraktur, spondyloarthritis, degenerative disc': 'scoliosis',
        'straight lumbal, fraktur, spondyloarthritis, degenerative disc': 'straight lumbal, fraktur, spondyloarthritis',

        'osteomyelitis, emfisema subkutis': 'emfisema subkutis',
        'pneumonia atau tb paru, efusi pleura': 'pneumonia atau tb paru',
        'pneumonia atau space occupying lesion, cardiomegaly': 'cardiomegaly',
        'pneumonia atau space occupying lesion': 'pneumonia',
        'tb paru, cardiomegaly, efusi pleura': 'tb paru',
        'pneumonia, dilatasi aorta': 'pneumonia',
        'pneumothorax, tb paru, cardiomegaly': 'cardiomegaly',
        'tb paru, efusi pleura': 'tb paru',
        'tb paru, emfisema': 'tb paru',
        'tb paru, atelektasis, efusi pleura': 'tb paru',
        'pneumothorax, pneumonia atau tb paru': 'pneumonia atau tb paru',
        'spondylolisthesis, straight lumbal, spondyloarthritis, hernia nucleus pulposus': 'hernia nucleus pulposus',
        'fraktur, hernia nucleus pulposus': 'hernia nucleus pulposus',
        'fraktur, arthritis': 'arthritis',
        'fraktur, osteoarthritis': 'osteoarthritis',
        'straight lumbal, scoliosis, spondyloarthritis, hernia nucleus pulposus': 'hernia nucleus pulposus',
        'straight lumbal, spondyloarthritis, degenerative disc': 'straight lumbal, spondyloarthritis',
        'straight lumbal, spondyloarthritis, fraktur, spondylolisthesis': 'straight lumbal, spondyloarthritis',
        'straight lumbal, spondyloarthritis, hernia nucleus pulposus': 'hernia nucleus pulposus',
        'straight lumbal, spondyloarthritis, scoliosis, degenerative disc': 'straight lumbal, spondyloarthritis',
        'straight lumbal, spondyloarthritis, spondylolisthesis, degenerative disc': 'straight lumbal, spondyloarthritis',
        'straight lumbal, verspondyloarthis': 'straight lumbal',
        'pneumonia, atelektasis, tracheal tube': 'pneumonia',
        'arthritis atau avascular necrosis': 'avascular necrosis',
        'artrithis sendi': 'arthritis',
        'pneumonia, atelektasis, efusi pleura': 'efusi pleura',
        'pneumonia, cardiomegaly, efusi pleura': 'cardiomegaly',
        'pneumonia, efusi pleura': 'efusi pleura',
        'pneumonia, space occupying lesion': 'space occupying lesion',
        'proses inflamasi': 'proses infeksi',
        'nefrolith dextra': 'nefrolith',
        'straight lumbal, fraktur, spondyloarthritis': 'straight lumbal, spondyloarthritis',
        'straight lumbal, spondylolisthesis': 'spondylolisthesis',
        'straight lumbal, fraktur, spondylolisthesis': 'spondylolisthesis',
        'spondyloarthritis, spondylolisthesis': 'spondyloarthritis',
        'spondylolisthesis, degenerative disc': 'degenerative disc',
        'spondyloarthritis, degenerative disc': 'degenerative disc',
        'straight lumbal, degenerative disc': 'degenerative disc',
        'pneumonia, emfisema': 'pneumonia',
        'fraktur, arthritis': 'arthritis',
        'metastasis, efusi pleura': 'efusi pleura',
        'aneurysma atau tumor mediastinum': 'tumor',


        'bronkhkpneumonia': 'bronkhopneumonia',
        'arthritis wrist joint dextra': 'arthritis',
        'disuse osteopoerosis atau osteomyelitis': 'disuse osteopoerosis',
        'ileus obstruksi': 'ileus',
        'ileus obstruktif': 'ileus',
        'ileus paralitik atau ileus obstruksi': 'ileus',
        'ileus obstruktif atau paralitik': 'ileus',
        'ileus atau paralitik': 'ileus',
        'ileus paralitik': 'ileus',
        'efusi pleura kiri': 'efusi pleura',


    }

    # Apply normalization
    for old_value, new_value in normalization_rules.items():
        conclusion = conclusion.replace(old_value, new_value)
    
    return conclusion

def determine_disease_type_kepala(conclusion):
    if conclusion == 'normal': 
        return conclusion
    if 'fraktur' in conclusion:
        conclusion = 'fraktur'
    if 'adenoid' in conclusion:
        conclusion = 'adenoid hypertrophy'
    if 'hipertrofi konka' in conclusion or 'pembesaran concha' in conclusion:
        conclusion = 'hipertrofi concha'
    if 'massa soft tissue' in conclusion:
        conclusion = 'tumor'
    return conclusion

def determine_disease_type_bno(conclusion):
    if conclusion == 'normal': 
            return conclusion
    if 'suspek' in conclusion:
        sentences = conclusion.split('.')
        modified_sentences = []
        for sentence in sentences:
            parts = sentence.split('suspek', 1)
            after_suspek = parts[1].strip() if len(parts) > 1 else ''
            before_suspek = parts[0].strip() 
            if 'dj stent' in before_suspek:
                modified_sentences.append('dj stent')
                break
            else:
                modified_sentences.append(after_suspek)
        conclusion = ''.join(modified_sentences)
    if conclusion.startswith('gambaran'):
        parts = conclusion.split('gambaran', 1)
        after_gambaran = parts[1].strip()
        conclusion = after_gambaran
    if 'dj stent' in conclusion:
        conclusion = 'dj stent'
    conclusion = conclusion.replace('.', '')
    conclusion = normalize_conclusion(conclusion)
    return conclusion

def determine_disease_type_abdomen(conclusion):
    if conclusion == 'normal': 
        return conclusion
    if conclusion == 'penebalan dinding pada colon ascendens suspek proses infeksi. bayangan opak pada abdomen kiri suspek nefrolith atau fecalith.':
        conclusion = 'nefrolith'
    if conclusion == 'bayangan lucent pada abdomen kiri suspek dilatasi gaster atau space occupying lesion.':
        conclusion = 'space occupying lesion'
    if conclusion == 'bayangan opak pada abdomen tengah suspek space occupying lesion.':
        conclusion = 'space occupying lesion'
    if conclusion == 'nefrolith dextra.':
        conclusion = 'nefrolith'
    if conclusion.startswith('gambaran'):
        parts = conclusion.split('gambaran', 1)
        after_gambaran = parts[1].strip()
        if 'obstruktif lbo' in after_gambaran:
            conclusion = after_gambaran.replace('obstruktif lbo', 'obstruktif').strip()
        else:
            conclusion = after_gambaran
    if conclusion.startswith('dilatasi'):
        if 'suspek' in conclusion:
            parts = conclusion.split('suspek', 1)
            after_suspek = parts[1].strip() if len(parts) > 1 else ''
            if after_suspek == 'dilatasi gaster atau space occupying lesion':
                conclusion = 'space occupying lesion'
            else:
                conclusion = after_suspek
    if conclusion.startswith('subileus'):
        conclusion = 'proses infeksi.'
    if conclusion.startswith('dj stent'):
        if conclusion.count('.') == 1:
            conclusion = 'dj stent'
        else:
            sentences = conclusion.split('.')
            second_sentences = sentences[1].strip()
            if 'sugestif' in second_sentences:
                parts = second_sentences.split('sugestif', 1)
                after_sugestif = parts[1].strip() if len(parts) > 1 else ''
                conclusion = after_sugestif
            if 'suspek' in second_sentences:
                parts = second_sentences.split('suspek', 1)
                after_suspek = parts[1].strip() if len(parts) > 1 else ''
                conclusion = after_suspek
    if conclusion.startswith('suspek'):
        if conclusion.count('.') == 1:
            parts = conclusion.split('suspek', 1)
            conclusion = ' '.join(parts[1].strip().split())
        else:
            sentences = conclusion.split('.')
            first_sentence = sentences[0].strip()
            if 'suspek' in first_sentence:
                parts = first_sentence.split('suspek', 1)
                conclusion = ' '.join(parts[1].strip().split())
    if conclusion.startswith('kalsifikasi'):
        conclusion = 'kalsifikasi'
    
    # Handling full stops
    conclusion = conclusion.replace('.', '')

    conclusion = normalize_conclusion(conclusion)
    print(conclusion)
    return conclusion

def determine_disease_type_ankle(conclusion):
    if conclusion == 'normal':
        return conclusion
    if 'fraktur' in conclusion:
        if 'internal fiksasi' in conclusion:
            conclusion = 'internal fiksasi, fraktur'
        else:
            conclusion = 'fraktur'
    if 'suspek' in conclusion:
        parts = conclusion.split('suspek')
        if len(parts) > 1:
            after_suspek = parts[1].strip()
            if after_suspek == 'proses inflamasi kronik.':
                conclusion = 'proses inflamasi'
    if 'sugestif' in conclusion:
        parts = conclusion.split('sugestif', 1)
        after_sugestif = parts[1].strip()
        conclusion = after_sugestif
        conclusion = conclusion.replace('.', '')
    conclusion = normalize_conclusion(conclusion)
    return conclusion

def determine_disease_type_antebrachii(conclusion):
    if conclusion == 'normal':
        return conclusion
    if 'suspek' in conclusion:
        parts = conclusion.split('suspek')
        if len(parts) > 1:
            after_suspek = parts[1].strip()
            conclusion = after_suspek
    if conclusion.startswith('fraktur'):
        conclusion = 'fraktur'
    if 'casting' in conclusion:
        conclusion = 'casting'
    if 'internal fiksasi' in conclusion:
        # conclusion = 'internal fiksasi'
        modified_sentence = []
        modified_sentence.append('internal fiksasi')
        if  'tampak fraktur' or 'tampak garis fraktur' or 'tampak samar garis fraktur' in conclusion:
            second_disease = 'fraktur'
            modified_sentence.append(second_disease)
        conclusion = ', '.join(modified_sentence)
    conclusion = conclusion.replace('.', '')
    return conclusion

def determine_disease_type_bahu(conclusion):
    if conclusion == 'normal':
        return conclusion
    if 'suspek' in conclusion:
        sentences = conclusion.split('.')
        modified_sentences = []
        for sentence in sentences:
            if 'suspek' in sentence:
                parts = sentence.split('suspek', 1)
                after_suspek = parts[1].strip() 
                if 'dislokasi' in after_suspek:
                    after_suspek = 'dislokasi'
                if 'tendinopathy' in after_suspek:
                    after_suspek = 'tendinopathy'
                modified_sentences.append(after_suspek)
            else:
                modified_sentences.append(sentence.strip())
        conclusion = ''.join(modified_sentences)
    if conclusion.startswith('fraktur'):
        if 'suspek' in conclusion:
            pass
        else:
            conclusion = 'fraktur'
    if conclusion.startswith('osteoarthritis'):
        conclusion = 'osteoarthritis'
    if 'internal fiksasi' in conclusion:
        conclusion = 'internal fiksasi'
    if conclusion == 'masih tampak non-union fraktur pada 1 3 tengah clavicula dengan angulasi dan displacement.':
        conclusion = 'fraktur'
    return conclusion

def determine_disease_type_cruris(conclusion):
    if conclusion == 'normal':
        return conclusion
    if 'internal fiksasi' in conclusion:
        # conclusion = 'internal fiksasi'
        modified_sentence = []
        modified_sentence.append('internal fiksasi')
        if  'tampak fraktur' or 'tampak garis fraktur' or 'tampak samar garis fraktur' in conclusion:
            second_disease = 'fraktur'
            modified_sentence.append(second_disease)
        conclusion = ', '.join(modified_sentence)
    if conclusion.startswith('fraktur'):
        conclusion = 'fraktur'
    if 'casting' in conclusion:
        conclusion = 'casting'
    if 'suspek' in conclusion:
        modified_sentences = []
        sentences = conclusion.split('.')
        first_sentence = sentences[0]
        second_sentence = sentences[1]
        if len(sentences) == 2:
            if 'fraktur' in first_sentence.split('suspek', 1)[1]:
                one_conclusion = 'fraktur'
            else:
                one_conclusion = first_sentence.split('suspek', 1)[1].strip().split()[0]
            modified_sentences.append(one_conclusion)
        elif len(sentences) == 3:
            if 'internal fiksasi' in sentences:
                pass
            if 'fraktur' in second_sentence:
                pass
            if 'suspek' in first_sentence:
                first_conclusion = first_sentence.split('suspek', 1)[1].strip().split()[0]
                modified_sentences.append(first_conclusion)
                # modified_sentences.append('fraktur')
        if len(modified_sentences) == 2:
            conclusion = ', '.join(modified_sentences)      
        else:
            conclusion = ''.join(modified_sentences)  
    if 'knee asthroplasty' in conclusion:
        conclusion = 'knee asthroplasty'
    return conclusion

def determine_disease_type_cubiti(conclusion):
    if conclusion == 'normal':
        return conclusion
    if conclusion.startswith('fraktur'):
        if 'suspek' in conclusion:
            pass
        else:
            conclusion = 'fraktur'
    if 'suspek' in conclusion:
        modified_sentences = []
        sentences = conclusion.split('.')
        first_sentence = sentences[0]
        second_sentence = sentences[1]
        if len(sentences) == 2:
            if 'fraktur' in first_sentence.split('suspek', 1)[1]:
                first_conclusion = 'fraktur'
            else:
                first_conclusion = first_sentence.split('suspek', 1)[1].strip()
                if 'osteomyelitis atau space occupying lesion' in first_conclusion:
                    first_conclusion = 'osteomyelitis'
            conclusion = first_conclusion
        elif len(sentences) == 3:
            if 'fraktur' in first_sentence:
                modified_sentences.append('fraktur')
            if 'suspek' in second_sentence:
                second_conclusion = second_sentence.split('suspek', 1)[1].strip()
                modified_sentences.append(second_conclusion)
            conclusion = ', '.join(modified_sentences)    
    if 'internal fiksasi' in conclusion:
        # conclusion = 'internal fiksasi'
        modified_sentence = []
        modified_sentence.append('internal fiksasi')
        if  'tampak fraktur' or 'tampak garis fraktur' or 'tampak samar garis fraktur' in conclusion:
            second_disease = 'fraktur'
            modified_sentence.append(second_disease)
        conclusion = ', '.join(modified_sentence)
    if 'casting' in conclusion:
        conclusion = 'casting'
    conclusion = normalize_conclusion(conclusion)
    return conclusion

def determine_disease_type_femur_genu(conclusion):
    if conclusion == 'normal':
        return conclusion
    sentences = conclusion.split('.')
    if 'fraktur' in conclusion:
        modified_sentences = []
        if len(sentences) == 2:
            conclusion = 'fraktur'
        elif len(sentences) == 3:
            first_sentence = sentences[0]
            second_sentence = sentences[1]
            if 'fraktur' in first_sentence:
                modified_sentences.append('fraktur')
            if 'osteoarthritis' in second_sentence:
                modified_sentences.append('osteoarthritis')
            conclusion = ', '.join(modified_sentences)
    if conclusion.startswith('osteoarthritis'):
        conclusion = 'osteoarthritis'
    if 'suspek' in conclusion:
        suspek_index = conclusion.find('suspek')
        conclusion = conclusion[suspek_index:].split()[1]
    conclusion = conclusion.replace('.', '')
    if 'internal fiksasi' in conclusion:
        # conclusion = 'internal fiksasi'
        modified_sentence = []
        modified_sentence.append('internal fiksasi')
        if  'tampak fraktur' or 'tampak garis fraktur' or 'tampak samar garis fraktur' in conclusion:
            second_disease = 'fraktur'
            modified_sentence.append(second_disease)
        conclusion = ', '.join(modified_sentence)
    if 'knee asthroplasty' in conclusion:
        conclusion = 'knee asthroplasty'
    conclusion = normalize_conclusion(conclusion)
    return conclusion

def determine_disease_type_hip_joint(conclusion):
    if conclusion == 'normal':
        return conclusion
    if 'suspek' in conclusion:
        parts = conclusion.split('suspek')
        after_suspek = parts[1].strip()
        disease_list = ['fraktur', 'avascular necrosis']
        for disease in disease_list:
            if disease in after_suspek:
                conclusion = disease
    if 'fraktur' in conclusion:
        if 'suspek' in conclusion:
            pass
        else:
            conclusion = 'fraktur'
    return conclusion

def determine_disease_type_humerus(conclusion):
    if conclusion == 'normal':
        return conclusion
    if 'fraktur' in conclusion:
        conclusion = 'fraktur'
    if 'internal fiksasi' in conclusion:
        # conclusion = 'internal fiksasi'
        modified_sentence = []
        modified_sentence.append('internal fiksasi')
        if  'tampak fraktur' or 'tampak garis fraktur' or 'tampak samar garis fraktur' in conclusion:
            second_disease = 'fraktur'
            modified_sentence.append(second_disease)
        conclusion = ', '.join(modified_sentence)
    return conclusion

def determine_disease_type_manus_os(conclusion):
    if conclusion == 'normal':
        return conclusion
    if 'suspek' in conclusion:
        suspek_index = conclusion.find('suspek')
        after_suspek = conclusion[suspek_index:].split()[1]
        disease_list = ['subluksasi', 'dislokasi', 'arthritis', 'mastoiditis', 'fraktur']
        for disease in disease_list:
            if disease in after_suspek:
                conclusion = disease
    if conclusion.startswith('fraktur'):
        conclusion = 'fraktur'
    if 'internal fiksasi' in conclusion:
        # conclusion = 'internal fiksasi'
        modified_sentence = []
        modified_sentence.append('internal fiksasi')
        if  'tampak fraktur' or 'tampak garis fraktur' or 'tampak samar garis fraktur' in conclusion:
            second_disease = 'fraktur'
            modified_sentence.append(second_disease)
        conclusion = ', '.join(modified_sentence)
    return conclusion

def determine_disease_type_vertebra(conclusion):
    if conclusion == 'normal':
        return conclusion
    if conclusion.startswith('straight lumbal'):
        parts = conclusion.split('.')
        first_conclusion = parts[0] # straight lumbal
        if len(parts) == 2:
            conclusion = first_conclusion
        if len(parts) == 3:
            modified_sentences = []
            disease_list = ['spondyloarthritis', 'spondylolisthesis', 'degenerative disc', 'verspondyloarthis']
            first_sentence = parts[0].strip()
            modified_sentences.append(first_sentence)
            second_sentence = parts[1].strip()
            for disease in disease_list:
                if disease in second_sentence:
                    second_conclusion = disease
                    modified_sentences.append(second_conclusion)
            conclusion = ', '.join(modified_sentences)
        if len(parts) == 4:
            modified_sentences = []
            first_sentence = parts[0].strip()
            modified_sentences.append(first_sentence)
            second_sentence = parts[1].strip()
            third_sentence = parts[2].strip()
            second_disease_list = ['spondyloarthritis', 'fraktur']
            third_disease_list = ['hernia nucleus pulposus', 'degenerative disc', 'spondylolisthesis']
            for second_disease in second_disease_list:
                if second_disease in second_sentence:
                    second_conclusion = second_disease
                    modified_sentences.append(second_conclusion)
            for third_disease in third_disease_list:
                if third_disease in third_sentence:
                    third_conclusion = third_disease
                    modified_sentences.append(third_conclusion)
            conclusion = ', '.join(modified_sentences)
        if len(parts) == 5:
            modified_sentences = []
            first_sentence = parts[0].strip()
            modified_sentences.append(first_sentence)
            second_sentence = parts[1].strip()
            third_sentence = parts[2].strip()
            fourth_sentence = parts[3].strip()
            second_disease_list = ['spondyloarthritis', 'scoliosis']
            third_disease_list = ['spondylolisthesis', 'spondyloarthritis', 'scoliosis', 'fraktur']
            fourth_disease_list = ['degenerative disc', 'hernia nucleus pulposus', 'spondylolisthesis']
            for second_disease in second_disease_list:
                if second_disease in second_sentence:
                    second_conclusion = second_disease
                    modified_sentences.append(second_conclusion)
            for third_disease in third_disease_list:
                if third_disease in third_sentence:
                    third_conclusion = third_disease
                    modified_sentences.append(third_conclusion)
            for fourth_disease in fourth_disease_list:
                if fourth_disease in fourth_sentence:
                    fourth_conclusion = fourth_disease
                    modified_sentences.append(fourth_conclusion)
            conclusion = ', '.join(modified_sentences)
    if conclusion.startswith('straight cervical'):
        parts = conclusion.split('.')
        if len(parts) == 2:
            conclusion = parts[0]  # straight cervical
        if len(parts) == 3:
            modified_sentences = []
            disease_list = ['spondyloarthritis']
            first_sentence = parts[0].strip()
            modified_sentences.append(first_sentence)
            second_sentence = parts[1].strip()
            for disease in disease_list:
                if disease in second_sentence:
                    second_conclusion = disease
                    modified_sentences.append(second_conclusion)
            conclusion = ', '.join(modified_sentences)
        if len(parts) == 4:
            modified_sentences = []
            first_sentence = parts[0].strip()
            modified_sentences.append(first_sentence)
            second_sentence = parts[1].strip()
            third_sentence = parts[2].strip()
            second_disease_list = ['spondyloarthritis']
            third_disease_list = ['kalsifikasi']
            for second_disease in second_disease_list:
                if second_disease in second_sentence:
                    second_conclusion = second_disease
                    modified_sentences.append(second_conclusion)
            for third_disease in third_disease_list:
                if third_disease in third_sentence:
                    third_conclusion = third_disease
                    modified_sentences.append(third_conclusion)
            conclusion = ', '.join(modified_sentences)
    if conclusion.startswith('fraktur'):
        parts = conclusion.split('.')
        first_sentence = 'fraktur'
        if len(parts) == 2:
            conclusion = first_sentence
        if len(parts) == 3:
            modified_sentences = []
            disease_list = ['hernia nucleus pulposus']
            modified_sentences.append(first_sentence)
            second_sentence = parts[1].strip()
            for disease in disease_list:
                if disease in second_sentence:
                    second_conclusion = disease
                    modified_sentences.append(second_conclusion)
            conclusion = ', '.join(modified_sentences)
        if len(parts) == 4:
            modified_sentences = []
            modified_sentences.append(first_sentence)
            second_sentence = parts[1].strip()
            third_sentence = parts[2].strip()
            second_disease_list = ['straight lumbal']
            third_disease_list = ['spondyloarthritis']
            fourth_disease_list = ['degenerative disc']
            for second_disease in second_disease_list:
                if second_disease in second_sentence:
                    second_conclusion = second_disease
                    modified_sentences.append(second_conclusion)
            for third_disease in third_disease_list:
                if third_disease in third_sentence:
                    third_conclusion = third_disease
                    modified_sentences.append(third_conclusion)
            conclusion = ', '.join(modified_sentences)
        if len(parts) == 5:
            modified_sentences = []
            modified_sentences.append(first_sentence)
            second_sentence = parts[1].strip()
            third_sentence = parts[2].strip()
            fourth_sentence = parts[3].strip()
            second_disease_list = ['straight lumbal']
            third_disease_list = ['spondyloarthritis']
            fourth_disease_list = ['degenerative disc']
            for second_disease in second_disease_list:
                if second_disease in second_sentence:
                    second_conclusion = second_disease
                    modified_sentences.append(second_conclusion)
            for third_disease in third_disease_list:
                if third_disease in third_sentence:
                    third_conclusion = third_disease
                    modified_sentences.append(third_conclusion)
            for fourth_disease in fourth_disease_list:
                if fourth_disease in fourth_sentence:
                    fourth_conclusion = fourth_disease
                    modified_sentences.append(fourth_conclusion)
            conclusion = ', '.join(modified_sentences)
    if conclusion.startswith('spondyloarthritis'):
        parts = conclusion.split('.')
        first_sentence = 'spondyloarthritis'
        if len(parts) == 2:
            conclusion = first_sentence
        if len(parts) == 3:
            modified_sentences = []
            disease_list = ['spondylolisthesis']
            modified_sentences.append(first_sentence)
            second_sentence = parts[1].strip()
            for disease in disease_list:
                if disease in second_sentence:
                    second_conclusion = disease
                    modified_sentences.append(second_conclusion)
            conclusion = ', '.join(modified_sentences)
    if conclusion.startswith('spondylolisthesis'):
        parts = conclusion.split('.')
        first_sentence = 'spondylolisthesis'
        if len(parts) == 3:
            modified_sentences = []
            disease_list = ['degenerative disc']
            modified_sentences.append(first_sentence)
            second_sentence = parts[1].strip()
            for disease in disease_list:
                if disease in second_sentence:
                    second_conclusion = disease
                    modified_sentences.append(second_conclusion)
            conclusion = ', '.join(modified_sentences)
    if conclusion.count('.') == 1:
        disease_list = ['hernia nucleus pulposus', 'degenerative disc', 'spondylolisthesis']
        for disease in disease_list:
            if disease in conclusion:
                conclusion = disease
    if conclusion.startswith('scoliosis'):
        parts = conclusion.split('.')
        first_sentence = 'scoliosis'
        if len(parts) == 2:
            conclusion = first_sentence
    if 'lateralisasi' in conclusion:
        conclusion = 'scoliosis'
    if conclusion.startswith('reverse spondylolisthesis'):
        conclusion = 'spondylolisthesis, hernia nucleus pulposus, straight lumbal, spondyloarthritis'
    if conclusion == 'kifosis thoracal. spondyloarthritis thoracolumbal. degenerative disc cv l2-l3.':
        conclusion = 'kifosis, spondyloarthritis, degenerative disc'
    if conclusion == 'scoliosis lumbal ke kanan. fraktur kompresi cv l3. straight lumbal. spondyloarthritis lumbal. penyempitan discus cv l4-l5 dan l5-s1 suspek degenerative disc disease.':
        conclusion = 'scoliosis, fraktur, straight lumbal, spondyloarthritis, degenerative disc'
    if conclusion == 'suspek spondylitis cv t10-cv t11. degenerative disc disease cv l4-l5.':
        conclusion = 'spondylitis, degenerative disc'
    if conclusion == 'spondylolisthesis cv l4- l5. multiple nefrolith dextra dan sinistra.':
        conclusion = 'spondylolisthesis, nefrolith'
    if conclusion == 'suspek spondylitis cv t12-cv l1.':
        conclusion = 'spondylitis'
    if conclusion == 'kompresi pada cv l2. spondyloarthritis lumbal. penyempitan discus cv l5-s1 suspek degenerative disc.':
        conclusion = 'spondyloarthritis, degenerative disc'
    if conclusion == 'kompresi pada cv l2 l3 dan l5. penyempitan discus cv l4-l5 dan l5-s1 suspek degenerative disc.':
        conclusion = 'degenerative disc'
    if conclusion == 'suspek fraktur kompresi cv l4. straight lumbal. spondyloarthritis lumbal.':
        conclusion = 'fraktur, straight lumbal, spondyloarthritis'
    if conclusion == 'fusi pada cv l3-l4 suspek spondylitis atau variasi normal.':
        conclusion= 'spondylitis'
    conclusion = normalize_conclusion(conclusion)
    return conclusion

def determine_disease_type_pedis(conclusion):
    if conclusion == 'normal':
        return conclusion 
    if 'soft tissue swelling' in conclusion:
        modified_sentence = []
        first_conclusion = 'proses infeksi'
        if 'kalsifikasi' in conclusion:
            second_conclusion = 'kalsifikasi'
            # modified_sentence.append(first_conclusion)
            modified_sentence.append(second_conclusion)
            conclusion = ''.join(modified_sentence)
        else:
            conclusion = first_conclusion
    if 'tophus gout' in conclusion:
        conclusion = 'artrithis sendi'
    if 'fraktur' in conclusion:
        conclusion = 'fraktur'
    if 'subluksasi' in conclusion:
        conclusion = 'subluksasi'
    if 'emfisema subkutis' in conclusion:
        parts = conclusion.split('.')
        modified_sentence = []
        if len(parts) == 2:
            conclusion = 'emfisema subkutis'
        if len(parts) == 3:
            first_conclusion = parts[0].strip()
            if 'suspek' in first_conclusion:
                suspek_index = first_conclusion.find('suspek')
                first_conclusion = first_conclusion[suspek_index:].split()[1]
                modified_sentence.append(first_conclusion)
                modified_sentence.append('emfisema subkutis')
                conclusion = ', '.join(modified_sentence)
            else:
                conclusion = 'emfisema subkutis'
        if len(parts) == 5:
            conclusion = 'emfisema subkutis'
    if 'internal fiksasi' in conclusion:
        conclusion = 'internal fiksasi'
    conclusion = normalize_conclusion(conclusion)
    return conclusion

def determine_disease_type_thorax(conclusion):
    if conclusion == 'normal':
        return conclusion
    if conclusion.startswith('pneumonia'):
        parts = conclusion.split('.')
        if len(parts) == 2:
            first_conclusion = parts[0].strip()
            if 'cavitas' in first_conclusion:
                first_conclusion = 'pneumonia atau tb paru'
            conclusion = first_conclusion
        if len(parts) == 3:
            modified_sentence = []
            first_conclusion = parts[0].strip()
            second_conclusion = parts[1].strip()
            disease_list = ['atelektasis', 'cardiomegaly', 'efusi pleura', 'dilatasi aorta', 'emfisema', 'space occupying lesion']
            if first_conclusion == 'pneumonia kanan':
                first_conclusion = 'pneumonia'
            modified_sentence.append(first_conclusion)
            for disease in disease_list:
                if disease in second_conclusion:
                    second_conclusion = disease
                    modified_sentence.append(second_conclusion)
            conclusion = ', '.join(modified_sentence)
        if len(parts) == 4:
            modified_sentence = []
            second_disease_list = ['cardiomegaly', 'atelektasis']
            third_disease_list = ['tracheal tube', 'efusi pleura']
            first_conclusion = parts[0].strip()
            second_conclusion = parts[1].strip()
            third_conclusion = parts[2].strip()
            modified_sentence.append(first_conclusion)
            for second_disease in second_disease_list:
                if second_disease in second_conclusion:
                    second_conclusion = second_disease
                    modified_sentence.append(second_conclusion)
            for third_disease in third_disease_list:
                if third_disease in third_conclusion:
                    third_conclusion = third_disease
                    modified_sentence.append(third_conclusion)
            conclusion = ', '.join(modified_sentence)
    if conclusion.startswith('tb paru'):
        parts = conclusion.split('.')
        if len(parts) == 2:
            conclusion = 'tb paru'
        if len(parts) == 3:
            modified_sentence = []
            first_conclusion = 'tb paru'
            second_conclusion = parts[1].strip()
            disease_list = ['atelektasis', 'efusi pleura', 'emfisema']
            modified_sentence.append(first_conclusion)
            for disease in disease_list:
                if disease in second_conclusion:
                    second_conclusion = disease
                    modified_sentence.append(second_conclusion)
            conclusion = ', '.join(modified_sentence)
        if len(parts) == 4:
            modified_sentence = []
            second_disease_list = ['atelektasis', 'cardiomegaly']
            third_disease_list = ['efusi pleura']
            first_conclusion = 'tb paru'
            second_conclusion = parts[1].strip()
            third_conclusion = parts[2].strip()
            modified_sentence.append(first_conclusion)
            for second_disease in second_disease_list:
                if second_disease in second_conclusion:
                    second_conclusion = second_disease
                    modified_sentence.append(second_conclusion)
            for third_disease in third_disease_list:
                if third_disease in third_conclusion:
                    third_conclusion = third_disease
                    modified_sentence.append(third_conclusion)
            conclusion = ', '.join(modified_sentence)
    single_disease_list = ['bronkhkpneumonia', 'cardiomegaly', 'dilatasi aorta', 'scoliosis', 'pleuropneumonia', 'hyaline membran disease', 'efusi pleura']
    if len(conclusion.split()) == 2:
        for disease in single_disease_list:
            if disease in conclusion:
                conclusion = disease
    if conclusion == 'efusi masif pleura kiri.':
        conclusion = 'efusi pleura'
    if conclusion == 'tak tampak kelainan radiologik pada jantung dan paru. elevasi diafragma kanan.':
        conclusion = 'elevasi diafragma'
    if conclusion == 'bronchitis dan emfisema pulmo.':
        conclusion = 'bronchitis, emfisema pulmo'
    if conclusion == 'pelebaran mediastinum suspek aneurysma aorta atau massa mediastinum.':
        conclusion = 'aneurysma atau tumor mediastinum'
    if conclusion == 'multiple nodul pada paru kanan suspek nodul metastasis. efusi pleura kiri.':
        conclusion = 'metastasis, efusi pleura'
    if conclusion.startswith('bayangan opak'):
        parts = conclusion.split('.')
        if len(parts) == 2:
            part = conclusion.split('suspek')
            after_suspek = part[1].strip()
            if after_suspek.startswith('round'):
                after_suspek = after_suspek[len('round'):].strip()
            conclusion = after_suspek
        if len(parts) == 3:
            disease_list = ['efusi pleura', 'cardiomegaly']
            modified_sentence = []
            first_conclusion = 'pneumonia atau space occupying lesion'
            second_conclusion = parts[1].strip()
            modified_sentence.append(first_conclusion)
            for disease in disease_list:
                if disease in second_conclusion:
                    second_conclusion = disease
                    modified_sentence.append(second_conclusion)
            conclusion = ', '.join(modified_sentence)
    if conclusion.startswith('pneumothorax'):
        parts = conclusion.split('.')
        if len(parts) == 3:
            first_conclusion = 'pneumothorax'
            second_conclusion = parts[1].strip()
            conclusion = first_conclusion + ', ' + second_conclusion
        if len(parts) == 4:
            modified_sentence = []
            second_disease_list = ['tb paru']
            third_disease_list = ['cardiomegaly']
            first_conclusion = 'pneumothorax'
            second_conclusion = parts[1].strip()
            third_conclusion = parts[2].strip()
            modified_sentence.append(first_conclusion)
            for second_disease in second_disease_list:
                if second_disease in second_conclusion:
                    second_conclusion = second_disease
                    modified_sentence.append(second_conclusion)
            for third_disease in third_disease_list:
                if third_disease in third_conclusion:
                    third_conclusion = third_disease
                    modified_sentence.append(third_conclusion)
            conclusion = ', '.join(modified_sentence)
    conclusion = conclusion.replace('.', '')
    conclusion = normalize_conclusion(conclusion)
    return conclusion

def determine_disease_type_pelvis(conclusion):
    if conclusion == 'normal':
        return conclusion
    if 'suspek' in conclusion:
        suspek_index = conclusion.find('suspek')
        after_suspek = conclusion[suspek_index + len('suspek'):].strip()
        disease_list = ['arthritis atau avascular necrosis', 'fraktur', 'avascular necrosis', 'osteomyelitis']
        for disease in disease_list:
            if disease in after_suspek:
                conclusion = disease
                break
    if conclusion.startswith('gambaran'):
        conclusion = conclusion[len('gambaran'):].strip()
        if 'striktur posterior urethra' in conclusion:
            conclusion = 'striktur urethra'
    if conclusion.startswith('fraktur'):
        conclusion = 'fraktur'
    if 'internal fiksasi' in conclusion:
        conclusion = 'internal fiksasi'
    if 'deformitas' in conclusion:
        conclusion = 'deformitas pada proximal femur'
    conclusion = normalize_conclusion(conclusion)
    return conclusion

def determine_disease_type_sinus(conclusion):
    if conclusion == 'normal':
        return conclusion
    if 'suspek' in conclusion:
        suspek_index = conclusion.find('suspek')
        conclusion = conclusion[suspek_index:].split()[1]
    if 'sinusitis' in conclusion:
        conclusion = 'sinusitis'
    return conclusion

def determine_disease_type_wrist(conclusion):
    if conclusion == 'normal':
        return conclusion
    if 'suspek' in conclusion:
        suspek_index = conclusion.find('suspek')
        after_suspek = conclusion[suspek_index + len('suspek'):].strip()
        disease_list = ['disuse osteopoerosis atau osteomyelitis', 'osteomyelitis', 'arthritis wrist joint dextra']
        for disease in disease_list:
            if disease in after_suspek:
                conclusion = disease
                break
    if 'fraktur' in conclusion:
        conclusion = 'fraktur'
    if 'osteoarthritis' in conclusion:
        conclusion = 'osteoarthritis'
    if 'casting' in conclusion:
        conclusion = 'casting'
    conclusion = conclusion.replace('.', '')
    conclusion = normalize_conclusion(conclusion)
    # print(conclusion)
    return conclusion
```

```python
# Create a dictionary mapping body parts to their corresponding functions
disease_type_functions = {
    'kepala': determine_disease_type_kepala,
    'BNO': determine_disease_type_bno,
    'abodemen_3_posisi': determine_disease_type_abdomen,
    'ankle': determine_disease_type_ankle,
    'antebrachii': determine_disease_type_antebrachii,
    'bahu': determine_disease_type_bahu,
    'cruris': determine_disease_type_cruris,
    'cubiti': determine_disease_type_cubiti,
    'femur': determine_disease_type_femur_genu,
    'genu': determine_disease_type_femur_genu,
    'hip_joint': determine_disease_type_hip_joint,
    'humerus': determine_disease_type_humerus,
    'manus': determine_disease_type_manus_os,
    'os_sacro_coccygeus': determine_disease_type_manus_os,
    'pelvis': determine_disease_type_pelvis,
    'sinus_paranasal': determine_disease_type_sinus,
    'wrist': determine_disease_type_wrist,
    'vertebra_cervical': determine_disease_type_vertebra,
    'vertebra_lumbosakral': determine_disease_type_vertebra,
    'vertebra_thoracolumbal': determine_disease_type_vertebra,
    'pedis': determine_disease_type_pedis,
    'thorax': determine_disease_type_thorax
}
```

```python
# Iterate through the DataFrame and apply the appropriate function based on the body part
for index, row in df.iterrows():
    body_part = row['body_part']
    conclusion = row['disease_type']
    
    # Get the function for the current body part
    determine_disease_type_function = disease_type_functions.get(body_part)
    
    # If a function is defined for the body part, use it to determine the disease type
    if determine_disease_type_function:
        df.at[index, 'disease_type'] = determine_disease_type_function(conclusion)

```
